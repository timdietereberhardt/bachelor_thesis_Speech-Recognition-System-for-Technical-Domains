# Optimization Variant 1 - Linear


```
import tensorflow as tf
import keras

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout, GRU

from keras.models import load_model
```

```
#Define the size of the smallest and biggest layer?
min_size_layer = 35
max_size_layer = 1000

#How much layers do you want to test?
num_layer = 10

#Set the start unit for the second last layer
start_unit = min_size_layer
max_start_unit = 300
dropout_rate = 0.1

#Rate how the 
steps = 10


def add_layer(size_list, dropout_rate):
    for layer in range(1,len(size_list)):
         unit_size =size_list[layer]

         model.add(GRU(unit_size, activation='relu', return_sequences= True))
         model.add(Dropout(dropout_rate))

def create_size_list(start_unit,size_layer,num_layer,steps):
    #Reset list
    size_list = []
    
    #Caclculate the layer size for each layer in a list
    for size in range(2,num_layer):
        size_list.append(round(start_unit+(size*size_layer)*steps))
    
    if num_layer > 4:
        size_list.reverse()

    return(size_list)


best = 0
num = 0

#Calculate how much models to create
num_models = round((max_size_layer-min_size_layer)/num_layer)

for size_layer in range(1,num_models):
        
        #Call func to create a list with the size of each layer
        size_list = create_size_list(start_unit,size_layer,num_layer,steps)
        print(size_list)

        
        #Create the start model
        model = Sequential()

        #Input Layer:
        input_layer_unit = size_list[0]

        #Change the unit size when its too big
        if input_layer_unit > max_size_layer:
            start_unit = start_unit*steps
            size_layer = 1

            #Call func
            size_list = create_size_list(start_unit,size_layer,num_layer,steps)
            input_layer_unit = size_list[0]

            #Check if the start_unit is under the limit
            if start_unit > max_start_unit:
             #   break

        model.add(GRU(input_layer_unit, input_shape = (20,16),activation='relu', return_sequences= True))
        model.add(Dropout(dropout_rate))

        #Add layers
        add_layer(size_list, dropout_rate)

        #Last Hidden Layer:
        model.add(GRU(start_unit, activation='relu',return_sequences= False))
        model.add(Dropout(dropout_rate))

        #Output Layer:
        model.add(Dense(9, activation='softmax'))

        #Stop the training when the model is overfitting --> Keras Callbacks
        callbacks_list = [
                    keras.callbacks.EarlyStopping(
                        monitor='val_loss',
                        patience=1,
                        mode='min', 
                        verbose=1,
                    )
        ]

        #Compile the model
        model.compile(optimizer='Adam',
                    loss='categorical_crossentropy',
                    metrics=['accuracy'])
        
        print("Modell", num," --> First Layer size: ",start_unit,"Last Layer Size:",input_layer_unit,"Number of layers:", num_layer )

        model.summary()
        #Train the model
        history = model.fit(x_seq_train,
                    y_train,
                    batch_size = 126,
                    epochs=100,
                    validation_data = (x_seq_val, y_val),
                    callbacks = callbacks_list)
        
        #Increase the Counter
        num += 1

        #Print out the best model
        if np.amax(history.history['val_acc']) > best:
              best = np.amax(history.history['val_acc'])
              print("Found new best model! Acc:",np.amax(history.history['acc']), "Val acc.", np.amax(history.history['val_acc']))
              model.summary()

              #Safe the best model
              model.save('best_opt_model.h5')
              model.save('/content/drive/My Drive/best_opt_model.h5')
        

print("Best model! Acc:",np.amax(history.history['acc']), "Val acc.", np.amax(history.history['val_acc']))
```

# Optimization Variant 2 - Random

```
import tensorflow as tf
import keras

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout, GRU

from keras.models import load_model
```

```
import random
#Define the size of the smallest and biggest layer?
min_size_layer = 35
max_size_layer = 1024

#How much layers do you want to test?
num_layer = 6

#Set the start unit for the second last layer
start_unit = min_size_layer
dropout_rate = 0.1


def add_layer(size_list, dropout_rate):
    for layer in range(1,len(size_list)):
         unit_size =size_list[layer]

         model.add(GRU(unit_size, activation='relu', return_sequences= True))
         model.add(Dropout(dropout_rate))

def create_size_list(start_unit,size_layer,num_layer,steps):
    #Reset list
    size_list = []
    size_list.append(start_unit)
    #Caclculate the layer size for each layer in a list
    for size in range(2,num_layer):
        size_list.append(random.randint(size_list[len(size_list)-1],size_list[len(size_list)-1]*2))
    
    if num_layer > 4:
        size_list.reverse()

    return(size_list)


best = 0
num = 0

#Calculate how much models to create
num_models = 100

for models in range(1,num_models):
        
        #Call func to create a list with the size of each layer
        size_list = create_size_list(start_unit,models,num_layer,steps)
        print(size_list)

        
        #Create the start model
        model = Sequential()

        #Input Layer:
        input_layer_unit = size_list[0]

        model.add(GRU(input_layer_unit, input_shape = (20,16),activation='relu', return_sequences= True))
        model.add(Dropout(dropout_rate))

        #Add layers
        add_layer(size_list, dropout_rate)

        #Last Hidden Layer:
        model.add(GRU(start_unit, activation='relu',return_sequences= False))
        model.add(Dropout(dropout_rate))

        #Output Layer:
        model.add(Dense(9, activation='softmax'))

        #Stop the training when the model is overfitting --> Keras Callbacks
        callbacks_list = [
                    keras.callbacks.EarlyStopping(
                        monitor='val_loss',
                        patience=1,
                        mode='min', 
                        verbose=1,
                    )
        ]

        #Compile the model
        model.compile(optimizer='Adam',
                    loss='categorical_crossentropy',
                    metrics=['accuracy'])
        
        print("Modell", num," --> First Layer size: ",start_unit,"Last Layer Size:",input_layer_unit,"Number of layers:", num_layer )

        model.summary()
        #Train the model
        history = model.fit(x_seq_train,
                    y_train,
                    batch_size = 126,
                    epochs=100,
                    validation_data = (x_seq_val, y_val),
                    callbacks = callbacks_list)
        
        #Increase the start_unit for more variability
        start_unit += 1

        #Print out the best model
        if np.amax(history.history['val_acc']) > best:
              best = np.amax(history.history['val_acc'])
              print("Found new best model! Acc:",np.amax(history.history['acc']), "Val acc.", np.amax(history.history['val_acc']))
              model.summary()

              #Safe the best model
              model.save('best_opt_model.h5')
              model.save('/content/drive/My Drive/best_opt_model.h5')
        

print("Best model! Acc:",np.amax(history.history['acc']), "Val acc.", np.amax(history.history['val_acc']))
```
